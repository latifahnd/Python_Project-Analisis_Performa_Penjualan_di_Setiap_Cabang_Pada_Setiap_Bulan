#### Dari pengecekan nama kolom untuk setiap dataset, diketahui bahwa semua dataset mempunyai list kolom yang sama, sehingga informasi dari kolom-kolom tersebut dapat digabungkan.

```python
df_tabel_retail=pd.concat([df_retail_data1, df_retail_data2, df_retail_data3, df_retail_data4])
print('Jumlah baris pada tabel retail:', df_tabel_retail.shape[0])
```

print(df_tabel_retail.info())

### Transformasi data
#### Pada langkah ini dilakukan tahapan untuk memastikan data yang memiliki item_price < 0 atau total_price < 0.

```python
data_cek=df_tabel_retail.loc[(df_tabel_retail['item_price'] < 0) | (df_tabel_retail['total_price'] < 0)]
```

```python
data_cek
```

#### Pada tabel diatas diketahui bahwa terdapat 2 data yang miliki item_price dan total_price yang tidak masuk akal, karena bernilai minus. Oleh karena itu, 2 data ini harus dihapus.

```python
if data_cek.shape[0] != 0:
    df_tabel_retail=df_tabel_retail.loc[(df_tabel_retail['item_price'] > 0) & (df_tabel_retail['total_price'] > 0)]
```

```python
data_cek=df_tabel_retail.loc[df_tabel_retail['order_id'] == 'undefined']
```

```python
data_cek
```

#### Pada tabel diatas diketahui bahwa terdapat 1 data yang miliki order_id bernilai undefined. Oleh karena itu, 1 data ini harus dihapus.

```python
if data_cek.shape[0] != 0:
    df_tabel_retail=df_tabel_retail.loc[df_tabel_retail['order_id'] != 'undefined']
```

#### Transformasi kolom order_id menjadi int64 dan kolom order_date menjadi datetime (Pandas).

```python
df_tabel_retail['order_id']=df_tabel_retail['order_id'].astype('int64')
df_tabel_retail['order_id'].dtypes
```

```python
df_tabel_retail['order_date']=pd.to_datetime(df_tabel_retail['order_date'])
df_tabel_retail['order_date'].dtypes
```

```python
df_tabel_retail.describe()
```

#### Melakukan filter terhadap 5 besar provinsi yang akan dianalisis (DKI Jakarta, Jawa Barat, Jawa, Timur, Jawa Tengah, dan Yogyakarta).

```python
provinsi5=['DKI Jakarta', 'Jawa Barat', 'Jawa Timur', 'Jawa Tengah', 'Yogyakarta']
df_tabel_retail=df_tabel_retail.loc[df_tabel_retail['province'].isin(provinsi5)]
```

#### Memastikan filter terhadap 5 besar provinsi berhasil.

```python
print(df_tabel_retail.province.unique())
```

#### Mengelompokkan data berdasarkan order_date dan province yang sudah difilter diatas dan menghitung order_unique_count, customer unique_count, product_unique_count, brand_unique_count, dan Gross Merchandise Volume (GMV) atau total_price untuk semua penjualan.

```python
provinsi_group=df_tabel_retail.groupby(['order_date','province']).agg({
    'order_id' : 'nunique',
    'customer_id' : 'nunique',
    'product_id' : 'nunique',
    'brand' : 'nunique',
    'total_price' : 'nunique',
    'total_price' : sum
})
```

#### Mengubah nama kolom menjadi 'order', 'customer', 'product', 'brand', dan 'GMV'.

```python
provinsi_group.columns=['order', 'customer', 'product', 'brand', 'GMV']
```

```python
provinsi_group
```

```python
provinsi_group.shape
```

```python
provinsi_group.head()
```

#### Melakukan unstack untuk mendapatkan order_date di bagian baris dan province di bagian column.

```python
unstack_provinsi_group=provinsi_group.unstack(level='province').fillna(0)
```

```python
unstack_provinsi_group
```

```python
unstack_provinsi_group.shape
```

```python
unstack_provinsi_group.head()
```

#### Visualisasi berupa grafik untuk masing-masing parameter perhitungan melalui slicing dan resampling untuk mendapatkan perhitungan rata-rata bulanan untuk masing-masing parameter perhitungan.

```python
#Membuat figure canvas dan axes untuk lima line plots.
figure,axes=plt.subplots(5, 1, figsize =(9,20))

#Slicing index
index=pd.IndexSlice['order', 'customer', 'product', 'brand', 'GMV']
for i, measurement in enumerate(provinsi_group.columns):
    #Slicing data untuk masing-masing measurement
    by_measurement=unstack_provinsi_group.loc[:, index[i]]
    #Melakukan resampling pada data untuk perhitungan rata-rata bulanan
    monthly_by_measurement_mean=by_measurement.resample('M').mean()
    #Plotting monthly_by_measurement_mean
    monthly_by_measurement_mean.plot(title='Average Daily ' + measurement + ' Size in Month View for 5 Provinces', ax=axes[i])
    axes[i].set_ylabel('Average ' + measurement + ' size')
    axes[i].set_xlabel('month')
    
plt.tight_layout()
plt.show()
```
